```python
# Make a scatter plot of sample size vs. unemployment rate

plt.scatter(grad_data['Sample_size'], grad_data['Unemployment_rate'])
plt.title('sample size vs. Unemployment_rate')
plt.xlabel('Sample Size')
plt.ylabel('Unemployment_rate')
```

```python
# Make a scatter plot of full time vs. median

plt.scatter(grad_data['Full_time_year_round'], grad_data['median'])
plt.title('Full_time_year_round vs. Median')
plt.xlabel('Full_time_year_round')
plt.ylabel('Median')
```
