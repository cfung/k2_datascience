## Assignment

Feel free to use native panda graphing methods.

- Make a scatter plot of sample size vs. median
- Make a scatter plot of sample size vs. unemployment rate
- Make a scatter plot of full time vs. median
- Make a scatter plot of momen vs. median

```python
# plt.scatter(df.preTestScore, df.postTestScore, s=df.age)
# ax.set_xlabel("x label")
# # y label
#plt.ylabel('Defender Size')

plt.scatter(grad_data['Sample_size'], grad_data['Median'])
plt.title('sample size vs. median')
plt.xlabel('Sample Size')
plt.ylabel('Median')
```

```python
# Make a scatter plot of sample size vs. unemployment rate

plt.scatter(grad_data['Sample_size'], grad_data['Unemployment_rate'])
plt.title('sample size vs. Unemployment_rate')
plt.xlabel('Sample Size')
plt.ylabel('Unemployment_rate')
```

```python
# Make a scatter plot of full time vs. median

plt.scatter(grad_data['Full_time_year_round'], grad_data['Median'])
plt.title('Full_time_year_round vs. Median')
plt.xlabel('Full_time_year_round')
plt.ylabel('Median')
```

```python
# Make a scatter plot of men vs. median

plt.scatter(grad_data['Men'], grad_data['Median'])
plt.title('Men vs. Median')
plt.xlabel('Men')
plt.ylabel('Median')
```

```python
# Make a scatter plot of women vs. median

plt.scatter(grad_data['Women'], grad_data['Median'])
plt.title('Women vs. Median')
plt.xlabel('Women')
plt.ylabel('Median')
```
