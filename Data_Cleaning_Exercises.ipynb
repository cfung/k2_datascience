<b>Exercise 3: Lets clean the 'attributes' column. The entries in this column are dictionaries. We need to do two things: </b>

<b> 1) Turn all the True or False in the dictionary to 1s and 0s.</b>

<b> 2) There are some entries within dictionaries that are dictionaries themselves, lets turn the whole entry into just one dictionary, for example if we have </b>

<b> '{'Accepts Credit Cards': True, 'Alcohol': 'none','Ambience': {'casual': False,'classy': False}}' </b>
<b> then turn it into</b>
<b> '{'Accepts Credit Cards':1, 'Alcohol_none': 1, 'Ambience_casual': 0, 'Ambience_classy': 0}'. </b>
<b> There might be other entries like {'Price Range': 1} where the values are numerical so we might want to change that into {'Price_Range_1': 1}.</b>

```python
def Flat(row):
    d = row['attributes']
    dl = {}
    
    for element in d.keys():
        if type(d[element]) == bool:
            dl[element] = int(d[element])
        
        elif type(d[element]) == str:
            label = '{}_{}'.format(element, d[element])
        
        elif type(d[element])== int:
            label = '{}_{}'.format(element, d[element])
            dl[label] = 1
        
        elif type(d[element]) == dict:
            for elem in d[element].keys():
                label = '{}_{}'.format(element, elem)
                dl[label] = int(d[element][elem])
        
        else:
            print (element, d[element])
    
    return dl
```

```python
data['attributes_clean'] = data.apply(lambda x: Flat(x), 1)
```
